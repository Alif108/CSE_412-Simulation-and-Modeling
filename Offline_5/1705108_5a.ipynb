```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Secretary Problem
n = 100
s_range = [1, 3, 5, 10]
```

```python
for s in s_range:
    success_rates = []
    
    for m in range(1, n):
        success_count = 0                                                                   # success count starts with 0
        for i in range(10000):

            candidates = np.random.sample(n)                                                # generate candidate list
            optimal_candidates = np.sort(candidates)[::-1][:s]                              # s number of max elements are optimal

            samples = np.random.choice(candidates, m, replace=False)                        # collect m random samples from candidates
            criterion = np.max(samples)                                                     # the max from the samples is the criterion

            final_cadidates = np.array([i for i in candidates if i not in samples])         # exclude the samples from the candidates

            chosen_candidate = -1                                                           # no candidate is chosen at first
            for candidate in final_cadidates:                                               # iterate over the final candidate list
                if candidate > criterion:                                                   # if a candidate is greater than criterion
                    chosen_candidate = candidate                                            # choose that candidate
                    break

            if chosen_candidate == -1:                                                      # if no candidate is chosen
                chosen_candidate = final_cadidates[-1]                                      # last candidate is chosen by default

            if chosen_candidate in optimal_candidates:                                      # if chosen candidate is in optimal candidate class
                success_count += 1                                                          # success!

        success_rates.append(success_count / 100)    # in percentage                        # count the successes and take the percentage : success_count * 100/ 10000

    plt.figure()
    plt.title('s = {}'.format(s))
    plt.xlabel('m')
    plt.ylabel('Success Rate')
    plt.plot(success_rates)
    plt.show()
```
