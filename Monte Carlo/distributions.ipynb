```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def normal_distribution(mean, variance):
    x = np.linspace(-5, 5, 100)
    y = 1 / np.sqrt(2 * np.pi * variance) * np.exp(-(x - mean) ** 2 / (2 * variance))
    plt.plot(x, y)
    plt.show()
    # return x, y 

# test the function
normal_distribution(0, 1)
```

```python
def chi_squared(mean, variance, df):
    x = np.linspace(0, 10, 100)
    y = (1 / (2 ** (df / 2) * np.sqrt(variance) * np.math.factorial(df - 1))) * x ** (df / 2 - 1) * np.exp(-x / 2)
    plt.plot(x, y, label = 'df = ' + str(df))
    plt.legend()
    plt.show()
    # return x, y

# test the function
chi_squared(0, 1, 2)
```

```python
# plot chi_squared distribution with different degrees of freedom in one figure
def chi_squared(mean, variance, df_range):
    x = np.linspace(0, 10, 100)
    
    for df in range(1, df_range):
        y = (1 / (2 ** (df / 2) * np.sqrt(variance) * np.math.factorial(df - 1))) * x ** (df / 2 - 1) * np.exp(-x / 2)
        plt.plot(x, y, label = 'df = ' + str(df))
        plt.legend()
    plt.show()
    # return x, y

# test the function
chi_squared(0, 1, 9)
```
